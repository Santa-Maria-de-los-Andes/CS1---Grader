## 🚀 Desafío Guiado — ¿Cuánto combustible necesita un cohete?

Este desafío te lleva paso a paso a través de una de las ecuaciones más importantes en ingeniería espacial: **la Ecuación de Tsiolkovsky**.

> *"La Tierra es la cuna de la humanidad, pero no se puede vivir en una cuna para siempre."*
> — Konstantin Tsiolkovsky, padre de la astronáutica

### El problema

SpaceX quiere lanzar un satélite al espacio. ¿Cuánto combustible necesita el cohete?

**Datos del cohete:**

| Variable | Valor | Descripción |
|----------|-------|-------------|
| `masa_carga` | 5,000 kg | Peso del satélite |
| `masa_seca` | 15,000 kg | Peso del cohete vacío |
| `Isp` | 300 s | Eficiencia del motor (impulso específico) |
| `g0` | 9.81 m/s² | Gravedad terrestre |
| `delta_v` | 7,800 m/s | Velocidad necesaria para alcanzar órbita |

Necesitaremos `import math` para usar la función exponencial.

**Paso 1** — Define las variables y calcula `masa_sin_combustible`

Primero sumamos el peso del satélite + peso del cohete vacío.

````python
### TU CODIGO AQUI — PASO 1 ###
import math

masa_carga = 5000     # kg — peso del satélite
masa_seca  = 15000    # kg — peso del cohete vacío
Isp        = 300      # s  — eficiencia del motor
g0         = 9.81     # m/s² — gravedad
delta_v    = 7800     # m/s — velocidad para orbitar

masa_sin_combustible = masa_carga + masa_seca

print(f"Masa sin combustible: {masa_sin_combustible:,} kg")
````

**Paso 2** — Calcula `masa_ratio` (razón de masas de Tsiolkovsky)

**Fórmula:** `masa_ratio = math.exp(delta_v / (Isp * g0))`

> Este número dice cuántas veces más pesado debe ser el cohete lleno comparado con vacío.
> Un valor de ~14 significa que el cohete lleno pesa 14 veces más que el cohete vacío.

````python
### TU CODIGO AQUI — PASO 2 ###
masa_ratio = math.exp(delta_v /(Isp * g0))   # usa: math.exp(delta_v / (Isp * g0))

print(f"Razón de masas: {masa_ratio:.4f}")
print(f"El cohete lleno pesa {masa_ratio:.1f}x más que vacío")
````

**Paso 3** — Calcula `masa_combustible`

**Fórmula:** `masa_combustible = masa_sin_combustible * (masa_ratio - 1)`

> Esto nos da el combustible necesario: la diferencia entre el cohete lleno y el cohete vacío.

````python
### TU CODIGO AQUI — PASO 3 ###
masa_combustible = 0.0   # usa: masa_sin_combustible * (masa_ratio - 1)

print(f"Combustible necesario: {masa_combustible:,.0f} kg")
print(f"¡Eso equivale a {masa_combustible/1000:.1f} toneladas!")
````

**Paso 4** — Calcula `fraccion_combustible` (% del peso total que es combustible)

**Fórmula:** `fraccion_combustible = masa_combustible / (masa_sin_combustible + masa_combustible) * 100`

> Los cohetes modernos son ≈ 85–95% combustible. ¿Tu resultado tiene sentido?

````python
### TU CODIGO AQUI — PASO 4 ###
fraccion_combustible = 0.0   # usa la fórmula de arriba

masa_total = masa_sin_combustible + masa_combustible
print(f"\n{'='*46}")
print(f"  ANÁLISIS DEL COHETE")
print(f"{'='*46}")
print(f"  Satélite (carga):    {masa_carga:>10,} kg")
print(f"  Cohete vacío:        {masa_seca:>10,} kg")
print(f"  Combustible:         {masa_combustible:>10,.0f} kg")
print(f"  MASA TOTAL:          {masa_total:>10,.0f} kg")
print(f"{'='*46}")
print(f"  El {fraccion_combustible:.1f}% del cohete es combustible!")
````

## ⚡ Reto 2 — Dilatación del Tiempo ⭐⭐⭐ (+6 pts)

Según Einstein, el tiempo pasa más lento para objetos en movimiento — **dilatación del tiempo**.

**Fórmula de Lorentz:**
```
gamma        = 1 / math.sqrt(1 - v**2 / c**2)
t_movimiento = t_reposo * gamma
```

**Variables de entrada:**
```python
c        = 299792458  # m/s — velocidad de la luz
t_reposo = 1.0        # segundo medido en reposo

v1 = 0.5 * c    # 50% de c
v2 = 0.9 * c    # 90% de c
v3 = 0.99 * c   # 99% de c
```

**Variables de salida** (calcula para cada velocidad):
`gamma_1`, `t_movimiento_1`, `diferencia_1`  (y lo mismo para v2 y v3)

**Valores esperados:**

| Velocidad | `gamma` | `t_movimiento` |
|-----------|---------|---------------|
| 50% c | ≈ 1.1547 | ≈ 1.1547 s |
| 90% c | ≈ 2.2942 | ≈ 2.2942 s |

## 🌍 Reto 3 — Ley de Gravitación Universal ⭐⭐⭐ (+6 pts)

### Aplicación real
La **NASA** usa esta ley para calcular trayectorias a Marte.
**SpaceX y Blue Origin** la usan para calcular el combustible necesario para escapar la gravedad.

### Tu misión
Newton descubrió que todo par de objetos con masa se atraen:

**Fórmula:** `F = G * (m1 * m2) / r**2`

Donde `G = 6.674e-11` N·m²/kg² (constante gravitacional universal).
Y la aceleración: `a = F / m` (Segunda Ley de Newton)

---

**Caso 1 — Tierra y Luna:**
```python
m_tierra      = 5.972e24    # kg
m_luna        = 7.342e22    # kg
r_tierra_luna = 384400000   # m
# → F_tierra_luna,  a_luna = F / m_luna
```
Esperado: `F ≈ 1.982×10²⁰ N`,  `a_luna ≈ 0.00270 m/s²`

---

**Caso 2 — Persona en la Tierra:**
```python
m_persona = 70        # kg
r_tierra  = 6371000   # m (radio terrestre)
# → F_persona,  a_persona = F / m_persona
```
Esperado: `F ≈ 686 N`,  `a_persona ≈ 9.81 m/s²`  ← ¡es la gravedad conocida!

---

**Caso 3 — Sol y Tierra:**
```python
m_sol        = 1.989e30   # kg
r_sol_tierra = 1.496e11   # m
# → F_sol_tierra,  a_tierra = F / m_tierra
```
Esperado: `F ≈ 3.54×10²² N`,  `a_tierra ≈ 0.00593 m/s²`

> 💡 Pista: usa notación científica `6.674e-11` (significa 6.674 × 10⁻¹¹)

````python
### TU CODIGO AQUI ###

G = 6.674e-11   # constante gravitacional universal

# ── Caso 1: Tierra y Luna ─────────────────────────────
m_tierra      = 5.972e24
m_luna        = 7.342e22
r_tierra_luna = 384400000.0   # metros

F_tierra_luna = 0.0   # G * m_tierra * m_luna / r_tierra_luna**2
a_luna        = 0.0   # F_tierra_luna / m_luna

# ── Caso 2: Persona en la Tierra ──────────────────────
m_persona = 70.0
r_tierra  = 6371000.0   # radio terrestre en metros

F_persona = 0.0   # G * m_tierra * m_persona / r_tierra**2
a_persona = 0.0   # F_persona / m_persona

# ── Caso 3: Sol y Tierra ──────────────────────────────
m_sol        = 1.989e30
r_sol_tierra = 1.496e11   # metros

F_sol_tierra = 0.0   # G * m_sol * m_tierra / r_sol_tierra**2
a_tierra     = 0.0   # F_sol_tierra / m_tierra

print("Caso 1 — Tierra y Luna:")
print(f"  F = {F_tierra_luna:.3e} N")
print(f"  a_luna = {a_luna:.5f} m/s²")
print("\nCaso 2 — Persona en la Tierra:")
print(f"  F = {F_persona:.1f} N")
print(f"  a_persona = {a_persona:.4f} m/s²  ← ¿reconoces este número?")
print("\nCaso 3 — Sol y Tierra:")
print(f"  F = {F_sol_tierra:.3e} N")
print(f"  a_tierra = {a_tierra:.5f} m/s²")
````

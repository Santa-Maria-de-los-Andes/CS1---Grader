## 1.2 — Tipos de Datos

Python tiene **tres tipos básicos** que usaremos siempre:

| Tipo | Qué guarda | Ejemplo | Cómo reconocerlo |
|------|-----------|---------|------------------|
| `int` | Número entero | `edad = 15` | Sin punto decimal, sin comillas |
| `float` | Número decimal | `precio = 4.50` | Tiene **punto** `.` decimal |
| `str` | Texto (string) | `nombre = "Ana"` | Entre **comillas** |
| `bool` | Verdadero/Falso | `activo = True` | `True` o `False` |

Puedes verificar el tipo con `type(variable)`.

````python
# 👀 OBSERVA — type() te dice el tipo de cada dato
print(type(15))        # <class 'int'>
print(type(4.50))      # <class 'float'>
print(type("Hola"))    # <class 'str'>
print(type(True))      # <class 'bool'>

#-------------------------------------------------

# ¿Cuál es la diferencia entre 15 y "15"?
a = 15
b = "15"
print(a + a)   # → 30  (suma numérica)
print(b + b)   # → 1515 (concatenación de texto!)
````

## 1.4 — Operadores Aritméticos

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|----------|
| `+` | Suma | `5 + 3` | `8` |
| `-` | Resta | `10 - 4` | `6` |
| `*` | Multiplicación | `6 * 7` | `42` |
| `/` | División | `20 / 4` | `5.0` ← siempre `float` |
| `**` | Potencia | `2 ** 3` | `8` |
| `%` | Módulo (resto) | `10 % 3` | `1` |

> ⚠️ **Los paréntesis importan:** `(2 + 3) * 4 = 20` pero `2 + 3 * 4 = 14`
> Python sigue las reglas de precedencia matemática (PEMDAS).

````python
# 👀 OBSERVA — operadores en acción (bodega del barrio)
pan    = 1.50
leche  = 4.20
huevos = 6.00
total  = pan + leche + huevos
print(f"Total bodega: {total:.2f} soles")
print(f"Con 10% desc: {total * 0.90:.2f} soles")
print(f"IGV (18%):    {total * 0.18:.2f} soles")
````

## 1.5 — f-strings

Los **f-strings** son la forma más limpia de mezclar texto y variables.

Pon `f` antes de las comillas y escribe las variables entre `{llaves}`:

```python
nombre = "Ana"
edad   = 16

print(f"Me llamo {nombre} y tengo {edad} años.")
# → Me llamo Ana y tengo 16 años.
```

**Formato de decimales:**
```python
precio = 9.999
print(f"Precio: {precio:.2f}")   # → Precio: 10.00  (2 decimales)
```

````python
# 👀 OBSERVA — f-strings en acción
artista  = "Rosalía"
año_disco = 2023
precio_entrada = 89.50

print(f"Artista: {artista}")
print(f"Año: {año_disco}")
print(f"Entrada: S/ {precio_entrada:.2f}")
print(f"{artista} tiene un concierto en {año_disco} — entrada: S/ {precio_entrada:.2f}")
````
